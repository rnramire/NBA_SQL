```python
q = '''SELECT COUNT(*)
FROM game
GROUP BY game_id
HAVING COUNT(*) > 1
'''
pd.read_sql_query(q,sql_connect)
#no duplicate games
```

```python
q = '''SELECT season_id, game_id, matchup_home, wl_home, fg_pct_home, ft_pct_home, pts_home,
fg_pct_away, ft_pct_away, pts_away
FROM game
WHERE matchup_home LIKE "BOS%"
'''
pd.read_sql_query(q,sql_connect)
#All Boston games in history
```

```python
q = '''SELECT COUNT(*) as wins
FROM game
WHERE matchup_home LIKE "%BOS" AND wl_home == 'W' 
'''
pd.read_sql_query(q,sql_connect)
```

```python
q = '''SELECT COUNT(*) as boston_wins
FROM game
WHERE matchup_home LIKE "%BOS vs. LAL%" AND wl_home == 'W' 
'''
pd.read_sql_query(q,sql_connect)
#number of home wins versus the lakers
```

```python
q = '''SELECT matchup_home, COUNT(*) as games
FROM game
GROUP BY matchup_home
HAVING games > 100
ORDER BY games DESC

'''
pd.read_sql_query(q,sql_connect)
```
